# Breakfast Cereal Analysis

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, r2_score
%matplotlib inline

data = pd.read_csv('data/Breakfast_Cereals.csv')
data_num = data.select_dtypes('number')
data.head()
```

# Data Overview
While the header table of the data can be useful for quickly understanding the structure of the dataset, a number of 1-variable visualizations can help us see the distribution of the various parameters, which will help to contextualize results and understand our models.

## 1-Variable Stats
At this point, apart from seeing a couple values in the table above, we don't know what to expect from our numerical data. By showing density plots of each one, we can understand some of the most important basic features at a glance: distribution, range, normality, outliers, etc.

```python
i = 0
font = {'fontsize': 8 }
for key in data_num:
    i+= 1
    plt.subplot(3,3,i)
    density = sns.kdeplot(data[key], shade=True, legend=False, color='b')
    plt.title(key, font, pad=2) # Format titles

plt.subplots_adjust(wspace=0.5, hspace=0.5) # Set margins for subplots

plt.show()
```

A few things are immediately apparent from these density plots. First, we notice that most of the parameters are close to normally distributed - this is useful because many simpler models, such as a linear regression, rely on the assumption that the data is normally distributed. As such, normal data is easier to analyze and find patterns in, so our models will likely be more accurate as a result.

However, one sticks out: the plot of vitamins is almost entirely centered on 25, with some outliers scattered around from 0 to 100. This data is not very useful for many models, and may actually be counterproductive for some. In fact, it might be more useful to understand vitamins as a categorical variable, considering cereals with 25 vitamins and cereals with more or less vitamins as two discrete categories of cereal. We'll look more into this later on.

## Principle Component Analysis
This is a complex algorithm that will randomly generate vectors in 9-space and determine the variation in the data along each vector. More concretely, we can use PCA to determine which of our numerical variables contribute the most to the variation in the data. This helps a lot in feature selection: parameters without much variation won't contribute to our models very much, so we can potentially improve their performance by removing them.

First we'll look at the component variances in a combined bar graph/cumulative frequency plot. This helps us determine which components account for the most variance and as such will be most helpful in identifying useful parameters.

```python
from sklearn.decomposition import PCA
from sklearn import preprocessing

normalized_data = pd.DataFrame(preprocessing.scale(data_num), columns=data_num.columns)

pca = PCA().fit(normalized_data)
a = np.arange(9)

ratios = pca.explained_variance_ratio_
cumulative_ratios = [np.sum(ratios[:i]) for i in range(1, len(ratios)+1)]

fig, ax1 = plt.subplots(figsize=(8,6))

color = 'tab:blue'
ax1.bar(a, pca.explained_variance_ratio_)
ax1.set_title("Fraction of Variance Explained by Component")
ax1.set_xlabel("Component")
ax1.set_ylabel("Fraction of Total Variance")
ax1.tick_params(axis='y', labelcolor=color)

ax2 = ax1.twinx()
color = 'tab:red'
ax2.plot(cumulative_ratios, scaley=False, color=color)
ax2.grid(True)
ax2.tick_params(axis='y', labelcolor=color)
ax2.set_ylim([0,1.05])
ax2.set_ylabel("Cumulative Variance Explained")

plt.show()
```

The first two components alone account for over 50% of the variance in the dataset, so to simplify the analysis, we'll just look at those two in the following table.

```python
components = pd.DataFrame(pca.components_, columns=data_num.columns)
components[:2]
```

This table gives us a lot of useful information. We can see that the first component includes mostly Calories, Fiber, and Potassium. The second component relies mostly on Fat, Carbohydrates, and Sugar. Note that Vitamins and Sodium have small coefficients for both components 0 and 1 - based on this, they probably won't play much of a role in our models.

## Confusion Matrix
However, this doesn't give us much of an idea of how exactly these parameters relate to one another. One way we can examine this is with a confusion matrix, which visualizes the correlation between each of the parameters in the dataset.

```python
from matplotlib import cm

corr = data_num.corr()
fig, ax = plt.subplots()
cb = ax.matshow(corr, cmap=cm.get_cmap(name="bwr"), vmin=-1.0)
fig.colorbar(cb)

ax.set_xticks((ax.get_xticks() + 1)[:9])
ax.set_yticks((ax.get_yticks() + 1)[:9])
plt.xticks(ax.get_xticks(), data_num.columns, rotation='vertical')
plt.yticks(ax.get_yticks(), data_num.columns)

plt.show()
```

A couple relationships are immediately obvious from this plot:
* Fiber - Calories
* Potassium - Calories
* Potassium - Fiber
* Sugar - Carbohydrates

We're going to have a look at these plots, and in order to do so we'll make a function to create a scatterplot with a linear regression line and an $r^2$ metric, so we don't write the same thing four times.

Another interesting note: the columns for Potassium and Vitamins indicate low correlation with all other variables, which corresponds nicely to the relationships we identified in the PCA.

```python
from sklearn.linear_model import LinearRegression

def lin_plot(xcol, ycol, axis):
    X = data[[xcol]]
    y = data[[ycol]]

    model = LinearRegression()
    model.fit(X, y)

    predictions = model.predict(X)
    score = model.score(X,y)
    r2 = r2_score(y, predictions)

    axis.scatter(X, y)
    axis.plot(X, predictions, 'k', label='$r^2 = %0.2f$' % r2)
    axis.set_xlabel(xcol)
    axis.set_ylabel(ycol)
    axis.legend(loc="upper right")
```

```python
fig, axes = plt.subplots(2,2, figsize=(8,8))

lin_plot('Fiber', 'Potassium', axes[0][0])
lin_plot('Sugar', 'Carbohydrates', axes[0][1])
lin_plot('Calories', 'Potassium', axes[1][0])
lin_plot('Fiber', 'Calories', axes[1][1])

plt.show()
```

We've definitely identified some interesting relationships here. As we go through some more models, we might expect to see these coming up again.

# Predicting a Value
In many cases, people looking to control their weight use a method known as "calories in, calories out," or CICO, as a simple way of controlling their diet. For these people, the caloric content of breakfast cereal would likely be an important factor in their buying decisions. Here we will train a model to predict the caloric content of cereals from the quantities of other nutrients.

As mentioned earlier, we're going to exclude the `Vitamins` and `Sodium` data from our model since it doesn't seem to be very helpful.

```python
from sklearn.linear_model import LinearRegression

target = "Calories"

X = data_num.drop([target, "Vitamins", "Sodium"], axis=1)
y = data_num[[target]]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
linModel = LinearRegression()
linModel.fit(X_train, y_train)

pred_linModel = linModel.predict(X_test)
score = linModel.score(X_test, y_test)
print("Linear model score: " + str(score))
```

The score here shows that the model is very accurate with its predictions - surprising considering the small sample size. We could expect this model to make useful predictions about the caloric content of cereals with just some nutrition information.

# Examining Vitamins
Early on, we saw in the density plots that the distribution of vitamins was particularly odd, with most of the data grouped around 25. This is a particularly interesting challenge - there doesn't seem to be a clear pattern in the data at first glance, so it should be interesting to see if our model is able to find one. Because we're predicting an ostensibly continuous variable, we'll first try using a linear regression model.

```python
from sklearn.linear_model import LinearRegression

X = data_num.drop("Vitamins", axis=1)
y = data_num[["Vitamins"]]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
linModel = LinearRegression()
linModel.fit(X_train, y_train)
pred_linModel = linModel.predict(X_test)

score = linModel.score(X_test, y_test)
print("Linear Regression Model Accuracy: {}".format(score))
```

Clearly the linear regression isn't able to effectively predict the number of vitamins in our cereals. Let's investigate some other models.

## Logistic Regression
Earlier we mentioned the possibility of evaluating vitamins as a categorical variable. Let's define our categories thus:

| Category | Condition |
|----|---|
| 0 | y $\neq$ 25 |
| 1 | y = 25 |

We'll try to train a logistic regression model to identify this binary classification.

```python
from sklearn.linear_model import LogisticRegression

X = data_num.drop("Vitamins", axis=1)
y = np.ravel(data_num[["Vitamins"]] == 25)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3)
logModel = LogisticRegression()
logModel.fit(X_train, y_train)
pred_logModel = logModel.predict(X_test)

score = accuracy_score(y_true=y_test, y_pred=pred_logModel)
print("Logistic Regression Model Accuracy: {}".format(score))
```

## Understanding the Logistic Model
The logistic model is around 75% accurate at predicting whether the cereal has exactly 25 vitamins! At first, this seems like an impressive success rate. However, note the composition of our target sample. 

```python
def get_props(y):
    return [sum(y) / len(y), (len(y) - sum(y)) / len(y)]

props = get_props(y)
fig, (ax1, ax2) = plt.subplots(1,2, subplot_kw={'aspect':'equal'})
ax1.pie(props, labels=["True", "False"], colors=['g','r'], autopct='%1.1f%%')
ax1.set_title("Original Proportions")

props_pred = get_props(pred_logModel)
ax2.pie(props_pred, labels=["True", "False"], colors=['g','r'], autopct='%1.1f%%')
ax2.set_title("Prediction Proportions")

plt.show()
```

Note that almost 75% of our sample is represented by cereals with 25 vitamins. As a result, our model is able to achieve a 73% success rate by just guessing True for all its results. We can see in the chart that the model seems to do almost exactly that: over 85% of its predictions are True. So, while in the aggregate the model seems to do well, we can't expect it to be any more accurate when categorizing a given cereal than just guessing that the cereal has 25 vitamins.

That being said, this comparison of proportions is a very imprecise way of understanding categorization accuracy. A better strategy would be to use an ROC analysis, which uses a table of probabilities generated by the model and varies a decision threshold to understand the model's tendency to correctly predict positive results. In essence, it allows us to understand the model's specificity against its sensitivity at various thresholds of strictness. An ideal model has an area under the curve of 1: it would correctly determine the categories of all inputs with a threshold infinitely close to 0. A worst-case model, random guessing, would score about 50% accuracy at any threshold. So, we expect an AUC between 0.5 and 1.0, with higher scores better.

```python
from sklearn.metrics import roc_curve, auc

predictions = logModel.predict_proba(X_test)

y_test_prob = np.array([[int(not bool), int(bool)] for bool in y_test])
y_score = predictions

n_classes = 2

fpr = dict()
tpr = dict()
roc_auc = dict()

for i in range(n_classes):
    fpr[i], tpr[i], _ = roc_curve(y_test_prob[:, i], y_score[:, i])
    roc_auc[i] = auc(fpr[i], tpr[i])

# Compute micro-average ROC curve and ROC area
fpr["micro"], tpr["micro"], _ = roc_curve(y_test_prob.ravel(), y_score.ravel())
roc_auc["micro"] = auc(fpr["micro"], tpr["micro"])
```

```python
plt.figure()
lw = 2
cat = 1
plt.plot(fpr[cat], tpr[cat], color='darkorange',
         lw=lw, label='ROC curve (area = %0.2f)' % roc_auc[cat])
plt.plot([0, 1], [0, 1], color='navy', lw=lw, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic')
plt.legend(loc="lower right")
plt.show()
```

One of the first things to note about this curve is its distinct un-curviness. Because this sample is so small, there's very little data left in the testing group. Consequently, the ROC curve comes out looking a bit coarse.

More importantly, though, the area under the curve is not much than a random guess. This shows much more robustly the conclusion that our logistic model isn't able to correctly categorize our data based on vitamins.

# Differentiating Brands
A conscientious but brand-loyal customer, an investor, a CEO... There are a number of reasons that someone would want to consider nutrition values organized by the brand of cereal. First let's see what manufacturers we have and how many samples we have from each one.

```python
X = data_num
y = data["Manufacturer"]
y.value_counts()
```

Interesting, looks like it's pretty much dominated by Kellogs and General Mills. We can visualize this information to make this a little more clear.

```python
ct = y.value_counts()

fig, (ax1, ax2) = plt.subplots(1,2, subplot_kw={'aspect':'equal'}, figsize=(20,5))
plt.title("Cereal Manufacturer Representation")
ax1.barh(ct.index, ct)
ax1.invert_yaxis()
ax2.pie(ct, labels=ct.index, autopct='%1.1f%%')

plt.show()
```

We can try to classify these using a K-Nearest-Neighbors model, which looks at the classes of the nearest points to try to guess what the class of the target should be.

```python
from sklearn.neighbors import KNeighborsClassifier


X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)

model = KNeighborsClassifier()
model.fit(X_train, y_train)
predictions = model.predict(X_test)
knn_score = accuracy_score(y_test, predictions)
print("KNN Score: {}".format(knn_score))
```

Often we can improve the accuracy of a KNN model by changing the value of $K$, the number of neighbors the model looks at to classify each point. In this case, we are limited in this regard by the small sample set: if we allow the model to look at more than about 10 items, it ends up examining almost half the data! We can try to optimize $K$ in this model, but we shouldn't expect too much from it.

```python
accuracy = []
max_k = 10

for k in range(1, max_k):
    model = KNeighborsClassifier(k)
    model.fit(X_train, y_train)
    predictions = model.predict(X_test)
    accuracy.append(accuracy_score(predictions, y_test))

plt.plot(range(1,max_k), accuracy)
plt.xlabel("K value")
plt.ylabel("Model Accuracy")
plt.title("Optimizing k")
plt.ylim([0,1])
plt.grid()
```

Except for some noise, the model's performance seems pretty much independent of the value of K. If we want to improve our accuracy on this dataset, we'll have to try a different algorithm.

## Decision Tree Model
A decision tree classifier uses the training data to generate a kind of flowchart based on the various parameters it is given, that ends with a particular class depending on those parameters.

```python
from sklearn.tree import DecisionTreeClassifier
model = DecisionTreeClassifier()
model.fit(X_train, y_train)
predictions = model.predict(X_test)
decision_tree_score = accuracy_score(y_test, predictions)
print("Decision Tree score: {}".format(decision_tree_score))
```

The decision tree is a bit better here, scoring over 50%, but that's still not particularly powerful. If want to eke out better performance from these algorithms, we would need to take a closer look at our dataset to try to understand what's going on with our classes. Unfortunately, with such a small dataset, some categories are going to have very little data on them, and this model won't be able to identify them.

In any case, we've come up with some interesting revelations about this data with regard to calories, vitamins, and some interesting relationships between fiber and potassium as well as fiber and calories.
